```python
## Read a Whole file



with open("example.txt",'r') as file:
    content=file.read()
    print(content)
```

```python
## Read a file line by line

with open('example.txt','r') as file:
    for line in file:
        print(line.strip()) # removes newline character 
```

```python
## Writing a file(Overwriting)

with open("example.txt",'w') as file:
        file.write("Hello World\n")
        file.write("This is a new line.")
```

```python
## Wrie a file without over writing 

with open('example.txt','a') as file:
    file.write("Hi welcome\n")
```

```python
## Writing a list of lines to a file

lines = ["first line \n","Seond line\n","Third line\n"]
with open("example.txt","a") as file:
    file.writelines(lines)
```

```python
##  Read the content from source text file and write to a destination text file

with open("example.txt",'r') as source_file:
    content = source_file.read()

with open("destination.txt",'w') as destination_file:
    destination_file.write(content)
```

```python
## Writing and then reading a file

with open("example.txt","w+") as file: # w+ used for reading and writing, if the file not created it will be created
    file.write("Hello World \n")
    file.write("This is a new line \n")

    ## Move the file cursor to the beginning
    file.seek(0)

    ## Read the content in the file
    content = file.read()
    print(content)
```
