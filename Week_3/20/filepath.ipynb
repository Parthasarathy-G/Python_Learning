```python
## Create a new directory
import os
new = "package"
os.mkdir(new)
print(f"Directory'{new}'create")
```

```python
## Listing files and directories

items = os.listdir(".")
print(items)
```

```python
## Joining paths

dir_name = "folder"

file_name="file.txt"

fullpath = os.path.join(os.getcwd(),dir_name,file_name)
print(fullpath)
```

```python
path = "example.txt"

if os.path.exists(path):
    print(f"The path {path} exists")

else:
    print(f"The path {path} does not exists")
```

```python
## Checking if a path is a file or directory

path = "example.bin"

if os.path.isfile(path):
    print("It is a file")
elif os.path.isdir(path):
    print("It is a dic")
else:
    print("Neither a file or directory")
```

```python
## Getting the absolute path

relative_path = "example.txt"
print(os.path.abspath(relative_path))
```
