```python
## function copy

def welcome():
    return "Welcome to python programming"

welcome()
```

```python
wel=welcome
print(wel())
del welcome
print(wel())
```

```python
## closures

def main(func):
    msg="welcome"
    def sub():
        print("Welcome to the advance python programming")
        print(msg)
        func()
        print("hi")

    return sub()

```

```python
main(print)
```

```python
## closures

def main(func,lst):
    msg="welcome"
    def sub():
        print("Welcome to the advance python programming")
        print(func(lst))

    return sub()

```

```python
main(len,[1,2,3])
```

```python
## Decorators

def main(func):
    msg="welcome"
    def sub():
        print("Welcome to the advance python programming")
        print(func)

    return sub()
```

```python
def code():
    print("Coimbatore")
```

```python
main(code)
```

```python
@main
def code():
 print("Coimbatore")
```

```python
## Decoarator

def decorator(func):
 def wrapper():
  print("Before")
  func()
  print("After")
 return wrapper
```

```python
@decorator
def hello():
    print("Hello")
    
```

```python
hello()
```
