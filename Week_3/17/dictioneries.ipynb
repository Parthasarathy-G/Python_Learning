```python
student = {"name":"Parthasarathy","Age":23,"Initial":"G"}
print(student)
```

```python
print(student['Age'])
```

```python
## Get method

print(student.get("Age"))
print(student.get("Last name","Not available"))
print(student)
```

```python
## Modifying Dictionary

print(student)

student["Age"] = 33
print(student)
```

```python
student["address"]="India" ## Added a new key and value
print(student)
```

```python
del student['Age'] ## Delete a key and value pair 

print(student)
```

```python
## Dictionary methods

keys = student.keys()
print(keys)
values = student.values()
print(values)
items = student.items()
print(items)
```

```python
## Shallow copy

student_copy = student
print(student)
print(student_copy)
```

```python
student["name"]="Sarathy"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## Shallow copy

student["name"] = "Parthasarathy"
print(student)
print(student_copy1)

```

```python
## Iterating over dictionaries
## You can use loops to iterate over dictionaries, keys, values or items

## Iterating over keys

for keys in student.keys():
    print(keys)

## Iterating over values

for keys in student.values():
    print(keys)
```

```python
## Iterate over key value pairs

for key, values in student.items():
    print(f"{key}:{values}")
```

```python
## Nested dictionaries

student = {
    "student1":{"Name":"Parthasarathy","age":24},
    "student2":{"Name":"Hari","age":23}
}
print(student)
```

```python
## Access the nested dictionaries elements

print(student["student1"]["Name"])
print(student["student2"]["Name"])
```

```python
##Iterating over nested dictionaries 

for studen_id,student_info in student.items():
    print(f"{studen_id}:{student_info}")
    for key,value in student_info.items():
        print(f"{key}:{value}")
```

```python
## Dictionary comphrehension

square = {x: x** 2 for x in range(5)}
print(square)
```

```python
## Conditional dictionary comprehension

even = {x:x**2 for x in range(10) if x%2==0}

print(even)
```

```python
##Practical examples

## use a dictionary to count th frequency of elements in list

numbers=[1,2,3,4,5,6,1,2,3,4,3,3,2,2,1,1]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number]+=1
    else:
        frequency[number] = 1     


print(frequency)
```

```python
## Merge 2 dictionaries into one

dict1 = {"a":1,"b":2}
dict2 = {"c":3,"d":4}

merged = {**dict1, **dict2}
print(merged)
```
