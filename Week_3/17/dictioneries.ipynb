```python
## Creating a dictionaries

empty_dic = {}
print(type(empty_dic))
```

```python
empty_dic = dict()
empty_dic
```

```python
student = {"name":"Parthasarathy","Age":23,"Initial":"G"}
print(student)
```

```python
print(student['Age'])
```

```python
## Get method

print(student.get("Age"))
print(student.get("Last name","Not available"))
print(student)
```

```python
## Modifying Dictionary

print(student)

student["Age"] = 33
print(student)
```

```python
student["address"]="India" ## Added a new key and value
print(student)
```

```python
del student['Age'] ## Delete a key and value pair 

print(student)
```

```python
## Dictionary methods

keys = student.keys()
print(keys)
values = student.values()
print(values)
items = student.items()
print(items)
```

```python
## Shallow copy

student_copy = student
print(student)
print(student_copy)
```

```python
student["name"]="Sarathy"
print(student)
print(student_copy)
```
