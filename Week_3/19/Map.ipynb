```python
numbers = [1,2,3,4,5]

list(map(lambda x:x ** 2,numbers))
```

```python
## Map multiple iterables

numbers1 = [1,2,3]
numbers2 = [4,5,6]

add = list(map(lambda x,y:x+y,numbers1,numbers2))

print(add)
```

```python
words = ["Apple","Mango","Banana"]

upper = list(map(str.upper,words))

print(upper)
```

```python
def get_name(person):
    return person["name"]

people = [
    {"name":"Partha","age":23},
    {"name":"Sarathy","age":24}
]

list(map(get_name,people))
```
